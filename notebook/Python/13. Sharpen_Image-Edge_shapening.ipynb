Image Sharpening = High intensity kernal + Gaussian blur

### Import Libraries

```python
import cv2
import numpy as np
import scipy.ndimage as scp
import matplotlib.pyplot as plt
```

### Read Image

```python
img = cv2.imread('Einstein.jpg', cv2.IMREAD_GRAYSCALE)
```

### Sharpening of Image (Blackbox)

```python
kernal_impulse = np.zeros((3,3))
kernal_impulse[1,1] = 2
blur_kernal = np.ones((3,3))/9
kernal = kernal_impulse - blur_kernal
```

```python
sharpen_image = scp.correlate(img, kernal)
```

```python
plt.figure(figsize=(20,18))
plt.subplot(1,2,1), plt.imshow(img, cmap='gray'), plt.title('Original')
plt.subplot(1,2,2), plt.imshow(sharpen_image, cmap='gray'), plt.title('Sharpen')
plt.show()
```

### Sharpening of Image (using inbuilt methods) 
In older days unsharp meant sharp because of negative photo films

```python
gaussian_3 = cv2.GaussianBlur(img, (9,9), 10)
unsharp_image = cv2.addWeighted(img, 1.5, gaussian_3, -0.5, 0)
plt.figure(figsize=(20,18))
plt.subplot(1,2,1), plt.imshow(img, cmap='gray'), plt.title('Original')
plt.subplot(1,2,2), plt.imshow(unsharp_image, cmap='gray'), plt.title('Sharpen')
plt.show()
```
