### Import Libraries

```python
import cv2
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
import skimage.util as ski
```

### Read Image

```python
image = cv2.imread('superman.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
print('shape:', image.shape)
```

### Gaussian noise

```python
Gausian_image = ski.random_noise(image, mode='gaussian', seed=128)
plt.figure(figsize=(15,12))
plt.subplot(1,2,1), plt.imshow(image, cmap='gray'), plt.title('Original')
plt.subplot(1,2,2), plt.imshow(Gausian_image, cmap='gray'), plt.title('Gaussian')
plt.show()
```

### Poisson Noise

```python
Poisson_image = ski.random_noise(image, mode='poisson', seed=128)
plt.figure(figsize=(15,12))
plt.subplot(1,2,1), plt.imshow(image, cmap='gray'), plt.title('Original')
plt.subplot(1,2,2), plt.imshow(Poisson_image, cmap='gray'), plt.title('Poisson')
plt.show()
```

### Salt and Pepper noise

```python
SnP_image = ski.random_noise(image, mode='s&p', seed=128)
plt.figure(figsize=(15,12))
plt.subplot(1,2,1), plt.imshow(image, cmap='gray'), plt.title('Original')
plt.subplot(1,2,2), plt.imshow(SnP_image, cmap='gray'), plt.title('S&P')
plt.show()
```

### Salt noise

```python
S_image = ski.random_noise(image, mode='salt', seed=128)
plt.figure(figsize=(15,12))
plt.subplot(1,2,1), plt.imshow(image, cmap='gray'), plt.title('Original')
plt.subplot(1,2,2), plt.imshow(S_image, cmap='gray'), plt.title('Salt')
plt.show()
```

### Pepper noise

```python
P_image = ski.random_noise(image, mode='pepper', seed=128)
plt.figure(figsize=(15,12))
plt.subplot(1,2,1), plt.imshow(image, cmap='gray'), plt.title('Original')
plt.subplot(1,2,2), plt.imshow(P_image, cmap='gray'), plt.title('pepper')
plt.show()
```

### Speckle Noise

```python
speckle_image = ski.random_noise(image, mode='speckle', seed=128)
plt.figure(figsize=(15,12))
plt.subplot(1,2,1), plt.imshow(image, cmap='gray'), plt.title('Original')
plt.subplot(1,2,2), plt.imshow(speckle_image, cmap='gray'), plt.title('speckle')
plt.show()
```
