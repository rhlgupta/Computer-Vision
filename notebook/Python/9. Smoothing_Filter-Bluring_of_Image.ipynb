These are Linear filters.

### Import Libraries

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

### Read Image

```python
img = cv2.imread('superman.jpg', cv2.IMREAD_GRAYSCALE)
ocv_img = cv2.imread('opencv.jpg', cv2.IMREAD_GRAYSCALE)
```

### Averaging Filtering

```python
plt.figure(figsize=(13,10))
A_blur = cv2.blur(img,(31,31)) # sigma = 31
ocv_A_blur = cv2.blur(ocv_img,(11,11)) # here sigma = 11 the blur and filter scale factor, and it can only be odd value
plt.subplot(221),plt.imshow(img, cmap='gray'),plt.title('Original')
plt.subplot(222),plt.imshow(A_blur, cmap='gray'),plt.title('Blurred')
plt.subplot(223),plt.imshow(ocv_img, cmap='gray'),plt.title('Original')
plt.subplot(224),plt.imshow(ocv_A_blur, cmap='gray'),plt.title('Smoothing Filter')
plt.show()
```

### Gaussian Filtering

```python
plt.figure(figsize=(13,10))
G_blur = cv2.GaussianBlur(img,(31,31),0)
ocv_G_blur = cv2.GaussianBlur(ocv_img,(11,11),0)
plt.subplot(221),plt.imshow(img, cmap='gray'),plt.title('Original')
plt.subplot(222),plt.imshow(G_blur, cmap='gray'),plt.title('Blurred')
plt.subplot(223),plt.imshow(ocv_img, cmap='gray'),plt.title('Original')
plt.subplot(224),plt.imshow(ocv_G_blur, cmap='gray'),plt.title('Smoothing Filter')
plt.show()
```

### Median Filtering

```python
plt.figure(figsize=(13,10))
median = cv2.medianBlur(img,31)
ocv_median = cv2.medianBlur(ocv_img,11)
plt.subplot(221),plt.imshow(img, cmap='gray'),plt.title('Original')
plt.subplot(222),plt.imshow(median, cmap='gray'),plt.title('Blurred')
plt.subplot(223),plt.imshow(ocv_img, cmap='gray'),plt.title('Original')
plt.subplot(224),plt.imshow(ocv_median,cmap='gray'),plt.title('Smoothing Filter')
plt.show()
```

### Bilateral Filtering

```python
plt.figure(figsize=(13,10))
B_blur = cv2.bilateralFilter(img,31,75,75)
ocv_B_blur = cv2.bilateralFilter(ocv_img,11,75,75)
plt.subplot(221),plt.imshow(img, cmap='gray'),plt.title('Original')
plt.subplot(222),plt.imshow(B_blur, cmap='gray'),plt.title('Blurred')
plt.subplot(223),plt.imshow(ocv_img, cmap='gray'),plt.title('Original')
plt.subplot(224),plt.imshow(ocv_B_blur,cmap='gray'),plt.title('Smoothin Filter')
plt.show()
```
