### Import Libraries

```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

### Read Images

```python
img1 = cv2.imread('demo1.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
img2 = cv2.imread('nature.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.figure(figsize=(20,15))
```

```python
plt.subplot(1,2,1)
```

```python
plt.imshow(img1, cmap='gray')
```

```python
plt.subplot(1,2,2)
```

```python
plt.imshow(img2, cmap='gray')
```

```python
plt.show()
```

### 1. Add Images

```python
plt.figure(figsize=(20,15))
```

```python
plt.subplot(1,3,1)
```

```python
img_sum = img1 + img2
```

plt.imshow(img_sum, cmap='gray')

```python
print('Image type:', img_sum.dtype)
```

```python
plt.title('1. A + B')
```

```python
plt.imshow(img_sum, cmap='gray')
```

### 2. Add Images using average

```python
plt.subplot(1,3,2)
```

```python
img_avg = (img1 + img2)/2
```

```python
plt.title('2. (A + B)/2')
```

```python
plt.imshow(img_avg, cmap='gray')
```

### 3. Add image using altered average

```python
plt.subplot(1,3,3)
```

```python
img_avg_alt = img1/2 + img2/2
```

```python
plt.title('3. A/2 + B/2')
```

```python
plt.imshow(img_avg_alt, cmap='gray')
```

```python
plt.show()
```
