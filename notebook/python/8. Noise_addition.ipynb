### Gaussian noise

```python
Impulse_noise_low = np.random.randn(590,670) * 32  # here sigma = 32; sigma refers to noise intensity
```

```python
Impulse_noise_high = np.random.randn(590,670) * 128 # here sigma = 128
```

```python
plt.figure(figsize=(10,8))
```

```python
plt.subplot(2,2,1)
```

```python
plt.title('Low Gaussian Noise (Sigma = 32)')
```

```python
plt.imshow(Impulse_noise_low, cmap='gray')
```

```python
plt.subplot(2,2,2)
plt.title('High Gaussian Noise (Sigma = 128)')
plt.imshow(Impulse_noise_high, cmap='gray')
```

```python
plt.subplot(2,2,3)
plt.title('Image + Low Gaussian Noise')
plt.imshow(image+Impulse_noise_low, cmap='gray')
```

```python
plt.subplot(2,2,4)
plt.title('Image + High Gaussian Noise')
plt.imshow(image+Impulse_noise_high, cmap='gray')
```

```python
plt.show()
```
